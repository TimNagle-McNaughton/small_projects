```python
### imports ###
import numpy as np
import plotly.express as px
import pandas as pd
import matplotlib.pyplot as plt
```

```python
### get the data ###
# read in csv
data = pd.read_csv("E:/rockfall/data.csv")

data["Aspect 20 degree centroids"]

directions = data["Aspect 20 degree centroids"]
directions_radians = 0.01745329*data["Aspect 20 degree centroids"]

HS_area = data["Percent in total HS area"]
HS_rockfalls = data["Count for HS rockfalls"]

LS_area = data["Pecent in total LS area"]
LS_rockfalls = data["Count for LS rockfalls"]
```

```python
def make_circular_plot(dataset, cmap, title, savename):
    fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(6, 6), constrained_layout=True)

    N = len(dataset)
    bottom = 0
    max_height = max(dataset)

    radii = dataset
    theta = directions_radians

    width = 2*np.pi/N

    ax = plt.subplot(111, polar=True)
    bars = ax.bar(theta, radii, width=width, bottom=bottom)

    if cmap == "magma":
        for r, bar in zip(radii, bars):
            bar.set_facecolor(plt.cm.magma(r / 10.))

    elif cmap == "viridis":
        for r, bar in zip(radii, bars):
            bar.set_facecolor(plt.cm.viridis(r / 10.))
    
    ax.set_title(str(title), va='bottom')
    ax.set_theta_zero_location("E")
    ax.set_rticks(list(range(0, int(max_height + 1), int(max_height/4))))
    plt.savefig("E:/rockfall/" + str(savename), dpi=300)
    plt.show()
```

```python
make_circular_plot(HS_area, "magma", "", "HS_Percent_nt")
make_circular_plot(HS_rockfalls, "viridis", "", "HS_Count_nt")
make_circular_plot(LS_area, "magma", "", "LS_Percent_nt")
make_circular_plot(LS_rockfalls, "viridis", "", "LS_Count_nt")
```

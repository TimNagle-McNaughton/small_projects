```python
# imports 
import random
import numpy as np
from scipy.spatial import Voronoi, voronoi_plot_2d
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
points = np.random.rand(80,2)
vor = Voronoi(points)

# set up the plot
fig, ax = plt.subplots(nrows=1, ncols=1, figsize=(6, 6), constrained_layout=True)

# set the plots
voronoi_plot_2d(vor, show_vertices=False, line_colors='k', line_width=1, line_alpha=1, point_size=3, ax=ax)

# turn off axes
plt.axis("off")

# save and show
plt.savefig(r"E:/Mars/networked_TARs/Figures/voronoi_100.png")
plt.show()
```
